```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import random as random
#import numpy as np
import pandas as pd
import time as time
import re as re
import csv as csv
from IPython.display import Image, display
import pickle
from collections import ChainMap
import itertools
from collections import Counter
import operator

#progress bar
def log_progress(sequence, every=None, size=None, name='Items'):
    from ipywidgets import IntProgress, HTML, VBox
    from IPython.display import display

    is_iterator = False
    if size is None:
        try:
            size = len(sequence)
        except TypeError:
            is_iterator = True
    if size is not None:
        if every is None:
            if size <= 200:
                every = 1
            else:
                every = int(size / 200)     # every 0.5%
    else:
        assert every is not None, 'sequence is iterator, set every'

    if is_iterator:
        progress = IntProgress(min=0, max=1, value=1)
        progress.bar_style = 'info'
    else:
        progress = IntProgress(min=0, max=size, value=0)
    label = HTML()
    box = VBox(children=[label, progress])
    display(box)

    index = 0
    try:
        for index, record in enumerate(sequence, 1):
            if index == 1 or index % every == 0:
                if is_iterator:
                    label.value = '{name}: {index} / ?'.format(
                        name=name,
                        index=index
                    )
                else:
                    progress.value = index
                    label.value = u'{name}: {index} / {size}'.format(
                        name=name,
                        index=index,
                        size=size
                    )
            yield record
    except:
        progress.bar_style = 'danger'
        raise
    else:
        progress.bar_style = 'success'
        progress.value = index
        label.value = "{name}: {index}".format(
            name=name,
            index=str(index or '?')
        )
```

```python
#FUNKCJE
#-----------

#funkcja która znajduje część wspólną tagów jeżeli time > time2 wsród sąsiadów noda a następnie zwraca wspólne tagi
def intersection(a,b,tg):
    inter_tag1 = list()
    inter_tag2 = list()
    tup = tuple()
    for (time,tag) in log_progress(a):
        if tag == tg:
            tup=(time,tag)
            inter_tag1.append(tup)
        for (time2, tag2) in b:
            if tag == tag2 and time > time2 and tag2 == tg:
                #print("1: ",time, tag)
                #print("2: ",time2, tag2)
                tup=(time2,tag2)
                inter_tag2.append(tup)
    print("1: ", inter_tag1, "\n2: ", inter_tag2)
    return tg

    
#szukanie najmniejszego timestampa dla zadanego tagu, jeżeli ta < tb t=99999999999999999999
def min_t(a,b,tg):
    inter_tag1 = list()
    inter_tag2 = list()
    for (time,tag) in a:
        if tag == tg:
            inter_tag1.append(time)
            #print("1: ",time, tag)
    for (time2, tag2) in b:
        if tag2 == tg:
            #print("2: ",time2, tag2)
            inter_tag2.append(time2)
    #print(min(inter_tag1), min(inter_tag2))
    if inter_tag1 and inter_tag2 and min(inter_tag1) > min(inter_tag2):
        return min(inter_tag2)
    else:
        return 99999999999999999999
    
def max_t(a,b,tg):
    inter_tag1 = list()
    inter_tag2 = list()
    for (time,tag) in a:
        if tag == tg:
            inter_tag1.append(time)
            #print("1: ",time, tag)
    for (time2, tag2) in b:
        if tag2 == tg:
            #print("2: ",time2, tag2)
            inter_tag2.append(time2)
    #print(min(inter_tag1), min(inter_tag2))
    if inter_tag1 and inter_tag2 and min(inter_tag1) < min(inter_tag2):
        return min(inter_tag2)
    else:
        return 99999999999999999999
    
#wybór n nodów z grafu g o tagu tag
def rep_node_set(n, g, tag):
    rand_nodes = list()

    lista = find_tag(tag)
    for i in range(0,n):
        rand_nodes.append(random.choice(lista))
        lista.remove(rand_nodes[i])
    rand_nodes.sort(key=lambda tup: tup[1])  # sorts in place
    return rand_nodes
        
        
#znajdowanie węzłów z zadanym tagiem bez powtórzeń
def find_tag(s):
    lista = list()
    elem = tuple()
    for x in g.nodes(data=True):
        find = 0
        for (time, tag) in x[1]['tt']:
            if s == tag and find == 0:
                    find = 1
                    elem = (x[0],time)
                    lista.append(elem)
    return lista

#długosc listy z powtorzeniami
def len_tag(s):
    lista = list()
    for x in log_progress(g.nodes(data=True)):
        for (time, tag) in x[1]['tt']:
            if s == tag:
                lista.append(x[0])
    return len(lista)

#długość listy bez powtórzeń
def len_tag2(s):
    lista = list()
    for x in log_progress(g.nodes(data=True)):
        find = 0
        for (time, tag) in x[1]['tt']:
            if s == tag and find == 0:
                find = 1
                lista.append(x[0])
    return len(lista)


#sprawdzenie sasiadów wierzchołka w przód lub tył
def check_neighbor(g, u, tag, direction):
    stop=0
    elem = tuple()
    x = u
    mini0 = (9999999999999999999,9999999999999999999)
    lista = list()    
    y=[n for n in g.neighbors(x)]
    for u2 in y:
        if direction == 1:
            score = min_t(g.node[u]['tt'], g.node[u2]['tt'], tag)
        elif direction == 0:
            score = max_t(g.node[u]['tt'], g.node[u2]['tt'], tag)
        elem = (score,u2)
        lista.append(elem)
    if lista:
        mini0 = min(lista)
    if mini0[0] < 9999999999999999999:
        mini = min(lista)
        x = mini
    else:
        mini = u
    out_tup = [i for i in lista if i[0] < 9999999999999999999]
    print(out_tup)
    return  x  

#znajdowanie n źródeł s0 
def find_s0(s,n):
    lista = list()
    min_list = list()
    #elem = tuple()
    for x in g.nodes(data=True):
        find = 0
        for (time, tag) in x[1]['tt']:
            if s == tag:
                    elem = (time,x[0])
                    lista.append(elem)
    for i in range(n):
        min_list.append(min(lista))
        lista.remove(min(lista))
    min_list2 = [i[1] for i in min_list]
    return min_list2

#znajdywanie źródła dla pojedynczego wierzchołka
def find_source(g, u, tag):
    stop=0
    elem = tuple()
    x = u
    mini0 = (9999999999999999999,9999999999999999999)
    hops = 0
    while stop==0:
        lista = list()    
        y=[n for n in g.neighbors(x)]
        for u2 in y:
            score = min_t(g.node[x]['tt'], g.node[u2]['tt'], tag)
            elem = (score,u2)
            lista.append(elem)
        if lista:
            mini0 = min(lista)
        #print(mini0)
        if mini0[0] < 9999999999999999999:
            mini = min(lista)
            x = mini[1]
            #stop=1
        else:
            mini = x
            stop = 1
    #print(lista)
    return  x

#znajdyawnie źródeł dla losowanego zbioru wierzchołków
def source(tag,n,g):
    lista = list()
    a = rep_node_set(n,g,tag)
    #print("rep_node_set: ",a)
    for (i, time) in log_progress(a):
        b =  find_source(g,i,tag)
        lista.append(b)
    return lista

#funkcja zwracajaca czy trafilismy dla zadanych param
# tg - tag, n-reprezentanci, k-liczba źródeł, d-ile najczesciej wyszukanych wybrać
def found(tg,n,k,d,g):
    print("Tag: ",tg)
    final_list = list()
    s0 = find_s0(tg,k)
    print("Sources: ",s0)
    suspected = source(tg,n,g)
    dicto = dict(Counter(suspected))
    for i in range(d):
        final = max(dicto.items(), key=operator.itemgetter(1))[0]
        final_list.append(final)
        del dicto[final]
    inter = set(s0).intersection(final_list)
    print("Founded: ",final_list)
    print("Interesection: ",inter )
    if inter:
        return 1
    else:
        return 0
```
