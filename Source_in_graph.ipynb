```python
import networkx as nx
import sys
import math
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
import random as random
import numpy as np
import pandas as pd
import time as time
import re as re
import csv as csv
from IPython.display import Image, display
import pickle
from collections import ChainMap
import itertools
from collections import Counter
import operator
import statistics

#progress bar
def log_progress(sequence, every=None, size=None, name='Items'):
    from ipywidgets import IntProgress, HTML, VBox
    from IPython.display import display

    is_iterator = False
    if size is None:
        try:
            size = len(sequence)
        except TypeError:
            is_iterator = True
    if size is not None:
        if every is None:
            if size <= 200:
                every = 1
            else:
                every = int(size / 200)     # every 0.5%
    else:
        assert every is not None, 'sequence is iterator, set every'

    if is_iterator:
        progress = IntProgress(min=0, max=1, value=1)
        progress.bar_style = 'info'
    else:
        progress = IntProgress(min=0, max=size, value=0)
    label = HTML()
    box = VBox(children=[label, progress])
    display(box)

    index = 0
    try:
        for index, record in enumerate(sequence, 1):
            if index == 1 or index % every == 0:
                if is_iterator:
                    label.value = '{name}: {index} / ?'.format(
                        name=name,
                        index=index
                    )
                else:
                    progress.value = index
                    label.value = u'{name}: {index} / {size}'.format(
                        name=name,
                        index=index,
                        size=size
                    )
            yield record
    except:
        progress.bar_style = 'danger'
        raise
    else:
        progress.bar_style = 'success'
        progress.value = index
        label.value = "{name}: {index}".format(
            name=name,
            index=str(index or '?')
        )
```

```python
#FUNKCJE
#-----------

#funkcja która znajduje część wspólną tagów jeżeli time > time2 wsród sąsiadów noda a następnie zwraca wspólne tagi
def intersection(a,b,tg):
    inter_tag1 = list()
    inter_tag2 = list()
    tup = tuple()
    for (time,tag) in log_progress(a):
        if tag == tg:
            tup=(time,tag)
            inter_tag1.append(tup)
        for (time2, tag2) in b:
            if tag == tag2 and time > time2 and tag2 == tg:
                #print("1: ",time, tag)
                #print("2: ",time2, tag2)
                tup=(time2,tag2)
                inter_tag2.append(tup)
    print("1: ", inter_tag1, "\n2: ", inter_tag2)
    return tg

#szukanie najmniejszego timestampa dla danego tagu w danym wierzchołku
def single_min_t(a,tg):
    tag1 = list()
    for (time,tag) in a:
        if tag == tg:
            tag1.append(time)
    if tag1:
        return min(tag1)
    else:
        return 99999999999999999999

#szukanie najmniejszego timestampa dla zadanego tagu w wierzchołku b w stosunku do a, jeżeli ta < tb t=99999999999999999999
def min_t(a,b,tg):
    inter_tag1 = list()
    inter_tag2 = list()
    for (time,tag) in a:
        if tag == tg:
            inter_tag1.append(time)
            #print("1: ",time, tag)
    for (time2, tag2) in b:
        if tag2 == tg:
            #print("2: ",time2, tag2)
            inter_tag2.append(time2)
    #print(min(inter_tag1), min(inter_tag2))
    if inter_tag1 and inter_tag2 and min(inter_tag1) > min(inter_tag2):
        return min(inter_tag2)
    else:
        return 99999999999999999999
    
def max_t(a,b,tg):
    inter_tag1 = list()
    inter_tag2 = list()
    for (time,tag) in a:
        if tag == tg:
            inter_tag1.append(time)
            #print("1: ",time, tag)
    for (time2, tag2) in b:
        if tag2 == tg:
            #print("2: ",time2, tag2)
            inter_tag2.append(time2)
    #print(min(inter_tag1), min(inter_tag2))
    if inter_tag1 and inter_tag2 and min(inter_tag1) < min(inter_tag2):
        return min(inter_tag2)
    else:
        return 99999999999999999999
    
#wybór n nodów z grafu g o tagu tag
def rep_node_set(n, g, tag):
    rand_nodes = list()

    lista = find_tag(tag)
    for i in range(0,n):
        rand_nodes.append(random.choice(lista))
        lista.remove(rand_nodes[i])
    rand_nodes.sort(key=lambda tup: tup[1])  # sorts in place
    return rand_nodes
        
        
#znajdowanie węzłów z zadanym tagiem bez powtórzeń
def find_tag(s):
    lista = list()
    elem = tuple()
    for x in g.nodes(data=True):
        find = 0
        for (time, tag) in x[1]['tt']:
            if s == tag and find == 0:
                    find = 1
                    elem = (x[0],time)
                    lista.append(elem)
    return lista

def find_tag2(s):
    lista = list()
    elem = tuple()
    for x in g.nodes(data=True):
        find = 0
        for (time, tag) in x[1]['tt']:
            if s == tag:
                    elem = (x[0],time)
                    lista.append(elem)
    return lista

#długosc listy z powtorzeniami
def len_tag(s):
    lista = list()
    for x in log_progress(g.nodes(data=True)):
        for (time, tag) in x[1]['tt']:
            if s == tag:
                lista.append(x[0])
    return len(lista)

#długość listy bez powtórzeń
def len_tag2(s):
    lista = list()
    for x in g.nodes(data=True):
        find = 0
        for (time, tag) in x[1]['tt']:
            if s == tag and find == 0:
                find = 1
                lista.append(x[0])
    return len(lista)


#sprawdzenie sasiadów wierzchołka w przód lub tył 1-przód, 0-tył
def check_neighbor(g, u, tag, direction):
    stop=0
    elem = tuple()
    x = u
    mini0 = (9999999999999999999,9999999999999999999)
    lista = list()    
    y=[n for n in g.neighbors(x)]
    for u2 in y:
        if direction == 1:
            score = min_t(g.node[u]['tt'], g.node[u2]['tt'], tag)
        elif direction == 0:
            score = max_t(g.node[u]['tt'], g.node[u2]['tt'], tag)
        elem = (score,u2)
        lista.append(elem)
    if lista:
        mini0 = min(lista)
    if mini0[0] < 9999999999999999999:
        mini = min(lista)
        x = mini
    else:
        mini = u
    out_tup = [i for i in lista if i[0] < 9999999999999999999]
    return  out_tup  

#znajdowanie n źródeł s0 
def find_s0(s,n):
    lista = list()
    min_list = list()
    #elem = tuple()
    for x in g.nodes(data=True):
        find = 0
        for (time, tag) in x[1]['tt']:
            if s == tag:
                    elem = (time,x[0])
                    lista.append(elem)
    for i in range(n):
        min_list.append(min(lista))
        lista.remove(min(lista))
    min_list2 = [i[1] for i in min_list]
    return min_list2

#znajdywanie źródła dla pojedynczego wierzchołka
def find_source(g, u, tag):
    stop=0
    elem = tuple()
    x = u
    mini0 = (9999999999999999999,9999999999999999999)
    hops = 0
    single_node = 0
    while stop==0:
        lista = list()    
        y=[n for n in g.neighbors(x)]
        for u2 in y:
            score = min_t(g.node[x]['tt'], g.node[u2]['tt'], tag)
            elem = (score,u2)
            lista.append(elem)
        if lista:
            mini0 = min(lista)
        if mini0[0] < 9999999999999999999:
            mini = min(lista)
            x = mini[1]
            hops=hops+1    
        else:
            mini = x
            times = single_min_t(g.node[u]['tt'],tag) - single_min_t(g.node[x]['tt'],tag)
            stop = 1
            if hops==0 and not check_neighbor(g, u, tag, 0):
                single_node = 1 
    return  x,hops,times,single_node

#znajdywanie teoretycznie prawdziwych źródeł dla losowanego zbioru wierzchołków
def source(tag,n,g):
    lista = list()
    a = rep_node_set(n,g,tag)
    #print("rep_node_set: ",a)
    for (i, time) in log_progress(a):
        b =  find_source(g,i,tag)
        lista.append(b[0])
    return lista

# tg - tag, n-obserwatorzy, k-liczba najwcześniejszych wystąpień, 
def found(tg,n,k,g):
    final_list = list()
    s0 = find_s0(tg,k)
    suspected = source(tg,n,g)
    print("Hasztag: ",tg)
    print("Najwcześniejszy wystąpienia: ",s0)
    print("Wylosowani obserwatorzy: ",suspected)
    inter = set(s0).intersection(suspected)
    print("Znalezione źródła: ",suspected)
    print("Część wspólna: ",inter )
    if inter:
        return 1
    else:
        return 0

def sources(tag,n,g):
    print("Hasztag: ",tg)
    node_list = list()
    hops_list = list()
    times_list = list()
    a_l = list()
    single_n = list()
    a = rep_node_set(n,g,tag)
    for (i, time) in a:
        b =  find_source(g,i,tag)
        node_list.append(b[0])
        hops_list.append(b[1])
        times_list.append(b[2])
        single_n.append(b[3])
        a_l.append(i)
    dicto = dict(Counter(node_list))
    node_list_final=[i for i in dicto.keys()]
    print("Wylosowani obserwatorzy: ",a_l)
    print("Znalezione źródła: ",node_list_final)
    return len(node_list_final), single_n.count(1), times_list, hops_list
    
        
       
```

```python
def research(tg,n,i,g):
    source_l = list()
    singles_l = list()
    times_ll = list()
    hops_ll = list()
    
    for y in log_progress(range(i)):
        a=sources(tg,n,g)
        source_l.append(a[0])
        singles_l.append(a[1])
        times_ll.append(a[2])
        hops_ll.append(a[3])
    
    times_l = [item for sublist in times_ll for item in sublist]
    hops_l = [item for sublist in hops_ll for item in sublist]
    
    dicto1 = dict(Counter(source_l))
    dicto2 = dict(Counter(singles_l))
    dicto3 = dict(Counter(times_l))
    dicto4 = dict(Counter(hops_l))
    
    zero_t = max(dicto3.items(), key=operator.itemgetter(1))[0]
    zero_h = max(dicto4.items(), key=operator.itemgetter(1))[0]
    del dicto3[zero_t]
    del dicto4[zero_h]
    
    times_no_0=[i for i in times_l if i != 0]
    
    if i>1:
        med_s = statistics.median(source_l)
        mean_s = statistics.mean(source_l)
        stdv_s = statistics.stdev(source_l)
        min_s = min(source_l)
        max_s = max(source_l)
        most_s = max(dicto1.items(), key=operator.itemgetter(1))[0]
        least_s = min(dicto1.items(), key=operator.itemgetter(1))[0]
    
        med_sin = statistics.median(singles_l)
        mean_sin = statistics.mean(singles_l)
        stdv_sin = statistics.stdev(singles_l)
        min_sin = min(singles_l)
        max_sin = max(singles_l)
        most_sin = max(dicto2.items(), key=operator.itemgetter(1))[0]
        least_sin = min(dicto2.items(), key=operator.itemgetter(1))[0]
    
    med_t = statistics.median(times_no_0)
    mean_t = statistics.mean(times_no_0)
    stdv_t = statistics.stdev(times_no_0)
    min_t = min(times_no_0)
    max_t = max(times_no_0)
    most_t = max(dicto3.items(), key=operator.itemgetter(1))[0]
    least_t = min(dicto3.items(), key=operator.itemgetter(1))[0]
    
    med_h = statistics.median(hops_l)
    mean_h = statistics.mean(hops_l)
    stdv_h = statistics.stdev(hops_l)
    min_h = min(hops_l)
    max_h = max(hops_l)
    most_h = max(dicto4.items(), key=operator.itemgetter(1))[0]
    least_h = min(dicto4.items(), key=operator.itemgetter(1))[0]
    old_stdout = sys.stdout
    sys.stdout = open(tg+'_'+str(n)+'.txt','wt')
    print("Hasztag: ",tg, "\nLiczba obserwatorów: ", n)
    if i>1:
        print("Statystyki źródeł:\n mediana: ",med_s, "\n średnia: ", mean_s, "\n odchylenie standardowe: ", stdv_s,"\n maksimum: ", max_s,"\n minimum: ", min_s,"\n najczęściej: ", most_s,"\n najrzadziej: ", least_s)
        print("Statystyki samotnych węzłów:\n mediana: ",med_sin, "\n średnia: ", mean_sin, "\n odchylenie standardowe: ", stdv_sin,"\n maksimum: ", max_sin,"\n minimum: ", min_sin, "\n najczęściej: ", most_sin,"\n najrzadziej: ", least_sin)
    print("Statystyki czasów:\n mediana: ",med_t, "\n średnia: ", mean_t, "\n odchylenie standardowe: ", stdv_t,"\n maksimum: ", max_t,"\n minimum: ", min_t,"\n najczęściej: ", most_t,"\n najrzadziej: ", least_t)
    print("Statystyki skoków:\n mediana: ",med_h, "\n średnia: ", mean_h, "\n odchylenie standardowe: ", stdv_h,"\n maksimum: ", max_h,"\n minimum: ", min_h,"\n najczęściej: ", most_h,"\n najrzadziej: ", least_h)
    if i==1:
        print("Samotne węzły: ", singles_l[0])
        print("Źródła: ", source_l[0])
    sys.stdout = old_stdout
    bins = np.linspace(math.ceil(min(times_no_0)), math.floor(max(times_no_0)),20) # fixed number of bins
    plt.hist(times_no_0,bins,edgecolor='black', linewidth=1.2)
    plt.title("CZASY ROZPRZESTRZENIANIA\n liczba obserwatorów="+str(n)+', hasztag='+tg)
    plt.xlabel('czas [s]')
    plt.ylabel('liczba wystąpień')
    #plt.show()
    plt.savefig(tg+'_'+str(n)+'_time.png', dpi = 300)
    plt.close()
    
    plt.hist(source_l,bins=np.arange(min(source_l), max(source_l)+2),edgecolor='black', linewidth=1.2,align='left')
    plt.title("ZNALEZIONE ŹRÓDŁA\n liczba obserwatorów="+str(n)+', hasztag='+tg)
    plt.xlabel('Liczba znalezionych źródeł')
    plt.ylabel('Liczba wystąpień')
    #plt.show()
    plt.savefig(tg+'_'+str(n)+'_sources.png', dpi = 300)
    plt.close()
    
    plt.hist(singles_l,bins=np.arange(min(singles_l), max(singles_l)+2),edgecolor='black', linewidth=1.2,align='left')
    plt.title("IZOLOWANE WĘZŁY\n liczba obserwatorów="+str(n)+', hasztag='+tg)
    plt.xlabel('Liczba znalezionych izolowanych węzłów')
    plt.ylabel('Liczba wystąpień')
    #plt.show()
    plt.savefig(tg+'_'+str(n)+'_singles.png', dpi = 300)
    plt.close()
    
    plt.hist(hops_l,bins=np.arange(min(hops_l), max(hops_l)+2),edgecolor='black', linewidth=1.2,align='left')
    plt.title("Długość ścieżki\n liczba obserwatorów="+str(n)+', hasztag='+tg)
    plt.xlabel('Długości ścieżki')
    plt.ylabel('Liczba wystąpień')
    #plt.show()
    plt.savefig(tg+'_'+str(n)+'_hops.png', dpi = 300)
    plt.close()
```
